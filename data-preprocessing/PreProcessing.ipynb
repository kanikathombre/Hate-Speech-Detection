# Stopwords removal for English & Hinglish

```python
nltk.download('stopwords')
from nltk.corpus import stopwords
```

```python
stop_words_english = set(stopwords.words('english'))
stop_words_hinglish = set(stopwords.words('hinglish'))

# Custom list of stopwords for Hinglish (if needed)
custom_stop_words = set(["the", "and", "is", "in", "of", "to", "that", "it", "he", "she","aur","fir","kaise","kyu","kyun","agar"])
```

```python
column = 'Comment'

def remove_stopwords(tokens):
    return [token for token in tokens if token.lower() not in stop_words_english and token.lower() not in stop_words_hinglish and token.lower() not in custom_stop_words]

data[column] = data[column].apply(remove_stopwords)
```

#Remove Special Charecters, Emojis & Numbers

```python
import re
```

```python
def remove_emojis_special_chars(text):
    emoji_pattern = re.compile("["
                               u"\U0001F600-\U0001F64F"  # emoticons
                               u"\U0001F300-\U0001F5FF"  # symbols & pictographs
                               u"\U0001F680-\U0001F6FF"  # transport & map symbols
                               u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                               u"\U00002500-\U00002BEF"  # chinese char
                               u"\U00002702-\U000027B0"
                               u"\U00002702-\U000027B0"
                               u"\U000024C2-\U0001F251"
                               u"\U0001f926-\U0001f937"
                               u"\U00010000-\U0010ffff"
                               u"\u2640-\u2642"
                               u"\u2600-\u2B55"
                               u"\u200d"
                               u"\u23cf"
                               u"\u23e9"
                               u"\u231a"
                               u"\ufe0f"  # dingbats
                               u"\u3030"
                               "]+", flags=re.UNICODE)
    text = emoji_pattern.sub(r'', str(text))

    text = re.sub(r'\d+', '', text)
    text = re.sub(r'[^\w\s]', '', text)

    return text
```

```python
data['Comment'] = data['Comment'].apply(remove_emojis_special_chars)
```

# Convert all tokens to lowercase

```python
from nltk.stem import PorterStemmer, WordNetLemmatizer
from nltk.tokenize import word_tokenize
nltk.download('wordnet')
```

```python
lemmatizer = WordNetLemmatizer()
column_name = 'Comment'
```

# Function for lemmatization

```python
def apply_lemmatization(text):
    tokens = word_tokenize(text)
    return ' '.join([lemmatizer.lemmatize(token) for token in tokens])
```

```python
data[column_name] = data[column_name].apply(apply_lemmatization)
```
